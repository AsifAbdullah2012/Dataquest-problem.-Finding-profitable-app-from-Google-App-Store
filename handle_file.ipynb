```python
from csv import reader

#this is the funtion for sclicing the rows and present them 
#first param: the list of lists, second param: first slicing index, third param: the end of the sclicing 
def explore_data(dataset, start, end, rows_and_columns):
    dataset_slice = dataset[start:end]    
    for row in dataset_slice:
        print(row)
        print('\n') # adds a new (empty) line after each row

    if rows_and_columns:
        print('Number of rows:', len(dataset))
        print('Number of columns:', len(dataset[0]))
        
#this is the code for reading the file and converting to lists 
# reading the Android applications 
opened_file = open('googleplaystore.csv', encoding="utf8")
read_file = reader(opened_file)
android = list(read_file)

    
open_file = open('AppleStore.csv', encoding = "utf8")
read_file = reader(open_file)
ios = list(read_file)

print('Here printing the Android Apps..........'+'\n')
explore_data(android,0,9,True)
print('Here printing the IOS Apps...........'+'\n')
explore_data(ios,0,5,True)



```

In this section we will delecte the duplicate rows of list items so that we may never fall into error while computation.


```python
dup_android = {}
dup_ios = {}
android_corrected = []
ios_corrected = []


for nam in android_not_wrong:
    val = nam[0]
    try:
        if val not in dup_android.keys():
            android_corrected.append(nam)
            dup_android[nam[0]] = 'found'
    except ValueError:
        print('error occured in {}'.format(nam))
for nam in ios_not_wrong:
    val = nam[0]
    try:
        if val not in dup_ios.keys():
            ios_corrected.append(nam)
            dup_ios[nam[0]] = 'found'
    except ValueError:
        print('error occured in {}'.format(nam))
android = android_corrected
ios = ios_corrected

print(len(android))
print(len(ios))
        
        
```

In the part below we have selected the english apps only, english apps containts characters that have ascii value from 0 to 127. So, that part was handled by a function called 'chec_eng'. So after selecting the english apps only, the number of of total apps have fallen. 

```python
def chec_eng(str):
    cnt = 0
    for char in str:
        if ord(char) > 127 or ord(char) < 0:
            cnt+=1
    if cnt > 3:
        return False
    else:
        return True
android_eng_app = []
ios_eng_app = []
for nam in android:
    str = nam[0]
    if chec_eng(str):
        android_eng_app.append(nam)
for nam in ios:
    str = nam[2]
    if chec_eng(str):
        ios_eng_app.append(nam)
print('total length of english android app is {}'.format(len(android_eng_app)))
print('total lenght of english ios app is {}'.format(len(ios_eng_app)))
tot = 0
for nam in android_eng_app:
    print(nam)
    tot = tot + 1
    if tot == 5:
        break
tot = 0
for nam in ios_eng_app:
    print(nam)
    tot = tot + 1
    if tot == 5:
        break

android = android_eng_app
ios = ios_eng_app
        
    
```
